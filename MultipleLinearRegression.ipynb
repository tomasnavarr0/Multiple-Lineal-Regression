```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("/content/houses_to_rent.csv")
ds=pd.DataFrame(data)
```

```python
ds.head()
```

```python
df=ds[["city","rooms","bathroom","floor","total"]]
df
```

```python
print(df.dtypes)
```

```python
df["floor"] = df["floor"].replace({'-':0}, regex=True)
df["floor"]=df["floor"].astype(int)
```

```python
df["total"] = df["total"].replace({'R':""}, regex=True)
df
```

```python
df["total"] = df["total"].replace({',':""}, regex=True)
df
```

```python
df["total"] = df["total"].replace({'$':""}, regex=True)
df
```

```python
df["total"]=df['total'].str[1:]
```

```python
df.head(50)
```

```python
city=df["city"].values
room=df["rooms"].values
bathroom=df["bathroom"].values
floor=df["floor"].values
total=df["total"].values
```

```python
x=np.array([city,room,bathroom,floor]).T
y=np.array(total)
```

```python
reg=LinearRegression()
```

```python
reg=reg.fit(x,y)
```

```python
Y_pred=reg.predict(x)
```

```python
r2=reg.score(x,y)
```

```python
print(f"error {r2}. los coeficientes son {reg.coef_}")
```
